* 39.6 Parsecs away from Earth
* Main sequence star with 1.5 $M_\odot$ and 4.9 $L_\odot$
* Debris disk as well as four large gas giants whose masses are each > 5 $M_J$ and radii 1.3 $R_J$
* Subsequent inferences about the central disk of smaller stars using LBTI
